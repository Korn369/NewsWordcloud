```python
from bs4 import BeautifulSoup
import requests
```

```python
words = []

for i in range(20):
    if(i == 0):
        url = 'https://thethaiger.com/news/national'
        page = requests.get(url)
        soup = BeautifulSoup(page.content, 'html.parser')
    else:
        url = url = f"https://thethaiger.com/news/national/page/{i+1}"
        page = requests.get(url)
        soup = BeautifulSoup(page.content, 'html.parser')
    headlines = soup.find_all("h2")
    for h in headlines:
        words.append(h.text.strip())
concatWord = ' '.join(words)
concatWord = concatWord.replace("Thailand","")
concatWord = concatWord.replace("Thai","")
        
```

```python
from wordcloud import WordCloud, STOPWORDS
import pandas as pd
import matplotlib.pyplot as plt
```

```python
stopwords= set(STOPWORDS)
new_words = []
new_stopwords=stopwords.union(new_words)
```

```python
# Size of Word Cloud
plt.rcParams["figure.figsize"] = (10,10)

# Make Wordcloud
wordcloud = WordCloud(max_font_size=100, max_words=40, background_color="black",stopwords=new_stopwords, colormap='gist_ncar_r').generate(concatWord)

# Plot Wordcloud
plt.plot()
plt.imshow(wordcloud, interpolation="none")
plt.axis("off")
plt.show()
```
